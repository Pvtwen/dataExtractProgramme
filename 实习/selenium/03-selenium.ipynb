### selenium软件测试的一个模块
selenium是一个用于Web应用程序测试的工具，它能够操控浏览器（Chrome、火狐、IE、Edge等）。

```python
import time
from selenium import webdriver
# 显示等待
from selenium.we
from selenium.webdriver.support import expected_conditions as EC
# 定位策略的包
from selenium.webdriver.common.by import Bybdriver.support.ui import WebDriverWait
# 期望的内容

# 创建谷歌浏览器对象
b = webdriver.Chrome()

# windows的selenium驱动
# b = webdriver.Chrome('C:\chromedriver.exe')

# 访问百度
#b.get(url='https://www.baidu.com/')
#time.sleep(2)
#b.get(url='https://taobao.com/')

b.get(url='https://101.qq.com/#/hero')

# 显示等待
# 显示等待就是根据指定的页面内容来确定什么时候结束等待
wait = WebDriverWait(b,100)
# 期望条件
A = EC.text_to_be_present_in_element(
    (By.XPATH,'//*[@id="app"]/div/div[3]/div/div/ul/li[1]/div/p'),
    '黑暗之女'
)
# 等待和指定条件相连接
# 等待，直到期望条件出现
wait.until(A)


# 打印页面源代码：page_source
print(b.page_source)

# 关闭浏览器
# b.close()
```
