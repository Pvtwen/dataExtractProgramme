```python
import numpy as np
import pandas as pd
df=pd.read_csv('2021年风向数据.csv')
df1=df[['日期','风向','风级']]
Direction=[]
Level=[]
for i in range(1,4):
    # 某一个月份前七天
    # [[],[],[],[]]
    direct=[]
    levels=[]
    str=f'2021-0{i}'
    df2=df1[(df1['日期']).str.startswith(str)]
    df3=df2[0:7]
    for j in range(0,7):
        direction=df3.iloc[j][1]
        level=df3.iloc[j][2]
        levels.append(level)
        direct.append(direction)
    Direction.append(direct)
    Level.append(levels)
print(Direction)
print(Level)
```

## dataframe格式的语法，因为在for循环中使用df.loc[i]获取某一行的数据会出错，卡了很久
### 获取i行的值
- df.iloc[i]
### 获取i行j列的值
- df.iloc[i][j]
### df切片
- df[i:j]
