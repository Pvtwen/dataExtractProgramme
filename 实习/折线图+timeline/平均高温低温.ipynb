```python
import numpy as np
import pandas as pd

# 改路径
df1=pd.read_csv('天气气温清洗后的数据.csv')
df2=df1[['日期','最高气温']]
# df3=df2[(df2['日期']).str.startswith('2021-01')]
# df3.sum()

# 一个大列表 每个小列表是一个年份
year=[]
for i in range(12,22):
    Day=[]    
    for j in range(1,10):
        date=f'20{i}-0{j}'
        df3=df2[(df2['日期']).str.startswith(date)]
        if len(df3)==0:
            continue
        Day.append(date)
    for k in range(10,13):
        date1=f'20{i}-{k}'
        df4=df2[(df2['日期']).str.startswith(date1)]
        if len(df4)==0:
            continue
        Day.append(date1)
    year.append(Day)
# 大列表 每个小列表是每个年份下属的每个月份的值
Max=[]
Min=[]
for i in range(12,22):
    # 1-9月
    AverMax=[]
    for j in range(1,10):
        date=f'20{i}-0{j}'
        df3=df2[(df2['日期']).str.startswith(date)]
        if len(df3)==0:
            continue
        #平均最高气温
        M=df3.mean()
        print(date+':')
        print(M)
        print('\n')
        AverMax.append(round(M[0],2))
#         print(date)
    # 10-12月
    for k in range(10,13):
        date1=f'20{i}-{k}'
        df4=df2[(df2['日期']).str.startswith(date1)]
        if len(df4)==0:
            continue
        M=df4.mean()
        print(date1+':')
        print(M)
        print('\n')
        AverMax.append(round(M[0],2))
#         print(date1)
    Max.append(AverMax)
print('--------------------')
Df2=df1[['日期','最低气温']]
for i in range(12,22):
    #1-9
    AverMin=[]
    for j in range(1,10):
        date=f'20{i}-0{j}'
        Df3=Df2[(Df2['日期']).str.startswith(date)]
        if len(Df3)==0:
            continue
        M=Df3.mean()
        print(date+':')
        print(M)
        print('\n')
        AverMin.append(round(M[0],2))
    for k in range(10,13):
        date1=f'20{i}-{k}'
        Df4=Df2[(Df2['日期']).str.startswith(date1)]
        if len(Df4)==0:
            continue
        M=Df4.mean()
        print(date1+':')
        print(M)
        print('\n')
        AverMin.append(round(M[0],2))
    Min.append(AverMin)
print('---------------------')
print(len(Day))
print(len(AverMax))
print(len(AverMin))

import pyecharts.options as opts
from pyecharts.charts import Line,Timeline

"""
Gallery 使用 pyecharts 1.1.0
参考地址: https://www.echartsjs.com/examples/editor.html?c=line-marker

目前无法实现的功能:

1、最低气温的最高值暂时无法和 Echarts 的示例完全复刻
"""
tl=Timeline()
x=0
for i in range(2012,2022):
    week_name_list = year[x]
    high_temperature = Max[x]
    low_temperature = Min[x]

    line=(
        Line(init_opts=opts.InitOpts(width="1600px", height="800px"))
        .add_xaxis(xaxis_data=week_name_list)
        .add_yaxis(
            series_name="最高气温",
            y_axis=high_temperature,
        )
        .add_yaxis(
            series_name="最低气温",
            y_axis=low_temperature,
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="未来一周气温变化", subtitle="纯属虚构"),
            tooltip_opts=opts.TooltipOpts(trigger="axis"),
            toolbox_opts=opts.ToolboxOpts(is_show=True),
            xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
        )
    )
    x+=1
    tl.add(line,"{}年".format(i))
tl.render("strikyLine.html")

# 最低气温
# str1='2012-01'
# df4=df2[(df2['日期']).str.startswith(str1)]
# print(len(df4))
```
